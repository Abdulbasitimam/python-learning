```python
#digital clock
from tkinter import Tk
from tkinter import Label
import time
import sys

root = Tk()
root.title('Clock')

def present_time():
  display_time = time.strftime('%I:%M:%S %P')
  digi_clock.config(text=display_time)
  digi_clock.after(200,present_time)

digi_clock= Lable(root, font= ('areal', 150), bg ='red', fg= 'black')
digi_clock.pack()
root.mainloop()
```

```python
import time
from IPython.display import clear_output

# This loop will run until you manually stop the cell
try:
    while True:
        # Get the current time
        # %I = 12-hour, %M = Minute, %S = Second, %p = AM/PM
        current_time = time.strftime('%I:%M:%S %p')

        # Clear the previous output so the clock stays on one line
        clear_output(wait=True)

        # Display the clock with some basic formatting
        print("==========================")
        print(f"  DIGITAL CLOCK: {current_time}")
        print("==========================")
        print("\n(Press the Stop button to exit)")

        # Wait for 1 second before updating
        time.sleep(1)
except KeyboardInterrupt:
    print("\nClock stopped.")
```
