```python
#resturant management

menu = {
    'Chai': 60,
    'Coffee': 100,
    'Noodle': 80,
    'Sandwich': 90,
    'Cake Rusk': 100
}
print(menu)
print('Welcome to our demo restaurant \nMenu Include Items')
print('Chai = Rs 60 \nCoffee = Rs 100 \nNoodle = Rs 80\nSandwich = Rs 90\nCake Rusk = Rs 100')

order_total = 0

item_1 = input('Enter name of item you want ')

if item_1 in menu:
  order_total += menu[item_1]
  print(f'Your item {item_1} has been added to order')

else:
  print(f'{item_1} is not available')

another_order= input('Do you want to add anything in the order? (Yes/No)')
if another_order == 'Yes':
  item_2 = input('Enter name of the item ')
  if item_2 in menu:
   order_total += menu[item_2]
   print(f'Your item {item_2} has been added to order')
  else:
    print(f'The given item {item_2} is not available')

print(f'The amount to pay of the order is {order_total}')

```
