```python
class person:
  def __init__(self, firstname, lastname, age, city, country):
    self.firstname= firstname
    self.lastname= lastname
    self.age= age
    self.city= city
    self.country= country
    self.skills= []

  def self_info(self):
    return print(f'{self.firstname} is a first and {self.lastname} last name while age is {self.age}, and country is {self.country} and city is {self.city}')

  def add_skills(self, skills):
    self.skills.append(skills)


p= person('abdul', 'basit', 23, 'karachi', 'pakistan')
p.self_info()
p.add_skills('html')
p.add_skills('java')
print(p.skills)



class Student(person):
    pass


s1 = Student('Eyob', 'Yetayeh', 30, 'Finland', 'Helsinki')
s2 = Student('Lidiya', 'Teklemariam', 28, 'Finland', 'Espoo')
print(s1.self_info())
s1.add_skills('JavaScript')
s1.add_skills('React')
s1.add_skills('Python')
print(s1.skills)

print(s2.self_info())
s2.add_skills('Organizing')
s2.add_skills('Marketing')
s2.add_skills('Digital Marketing')
print(s2.skills)

```
