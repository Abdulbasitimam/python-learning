```python
from datetime import datetime
new= datetime(2020,1,1)
day= new.day
month= new.month
year= new.year
hour= new.hour
minute= new.minute
print(f'{year}, {month}, {day}, {hour}:{minute}')

now= datetime.now()
t= now.strftime("%H:%M:%S")
print('Time:', t)
t_one= now.strftime("%y/%m/%d %h:%m:%s")
print(t_one)
```
